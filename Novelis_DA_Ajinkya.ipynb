```python
cleandf = pd.read_csv(r'/content/drive/MyDrive/data/clean.csv')
```

```python
cleandf
```

```python
cleandf = cleandf.drop('Percent',axis = 1)
cleandf = cleandf.drop('perc_min',axis = 1)
cleandf = cleandf.drop('perc_max',axis = 1)
cleandf = cleandf.drop('Cycle',axis = 1)
cleandf = cleandf.drop('Period',axis = 1)
```

```python
cleandf
```

```python
train = cleandf.sample(frac=0.8,random_state=57)
test = cleandf.drop(train.index)
```

```python
x_train = train.loc[:,train.columns != 'Class']
x_test = test.loc[:,test.columns!= 'Class']
y_train = train.loc[:,'Class']
y_test = test.loc[:,'Class']
```

```python
logreg = LogisticRegression(max_iter=10000)
```

```python
logreg.fit(x_train,y_train)
```

```python
logreg.classes_
```

```python
logreg.intercept_
```

```python
logreg.coef_
```

```python
logreg.predict(x_test)
```

```python
logreg.score(x_test,y_test)
```

```python
confusion_matrix(y_test,logreg.predict(x_test))
```

```python
rf = RandomForestClassifier(n_estimators=10000,random_state=57)
```

```python
rf.fit(x_train,y_train)
```

```python
rfpred = rf.predict(x_test)
```

```python
rf_acc = accuracy_score(y_test,rfpred)
```
